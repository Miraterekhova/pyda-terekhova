## Задание 1

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:  

1. `car_id = 'A222BC96'`

Результат: 
`Номер A222BС валиден. Регион: 96`

2. `car_id = 'АБ22ВВ193'`

Результат: 
`Номер не валиден`

````python
import re
````

````python
car_id = 'A222BC96'
````

````python
def valid_number(car_id):
    pattern = r'(\w{1}\d{3}\w{2})(\d{2,3})' # задаю маску
    if pattern:
        group_num = re.search(r'(\w{1}\d{3}\w{2})(\d{2,3})', car_id) # разделяю по группам
        print('Номер', {group_num.group(1)}, 'валиден. Регион:',{group_num.group(2)})
    else:
        print('Номер не валиден')
        
valid_number('A222BC96')
````

Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

Пример работы программы:  

`some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений.'`

Результат:  
`Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.`

````python
some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений.'
````

````python
def del_repeats(some_string):
    with_reapits = re.sub(r'\b(\w+)\s+\1\b', r'\1', some_string) 
    no_reapits = re.sub(r'\b(\w+)\s+\1\b', r'\1', with_reapits)
    return no_reapits
````

````python
del_repeats(some_string)
````

## Задание 3

Напишите функцию, которая будет возвращать акроним по переданной в нее строке со словами.

Примеры работы программы:  

1. `some_words = 'Информационные технологии'`

Результат: 
`ИТ`

2. `some_words = 'Near Field Communication'`

Результат: 
`NFC`

````python
some_words = 'Near Field Communication'
````

````python
def acronims(some_words):
    word = re.findall(r'[А-ЯЁа-яA-Za-z]\w+', some_words) # шаблон, который ищет все буквы в английском и русском алфавите
    res = []
    for acronim in word:
        a = acronim[0].upper() # цикл проходится по элементам во фразе(в данном случае по словам), возводит в верхний регистр и в каждом слове обращается к 0ому индексу
        res.append(a)  
        res = [''.join(res)] #соединяет буквы
    return res
````

````python
acronims(some_words)
````

## Задание 4

Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

Пример работы программы:  

`emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']`

Результат: 
```
gmail.com: 2
test.in: 1 
ya.ru: 2 
mail.ru: 1
```

````python
emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']
````

````python
def count_email(emails):
    result = []

    for mail in emails:
        domen = re.findall(r'(?:\w+)@([a-z]+\.[a-z]{2,3})',mail) #цикл, проходится по каждому элементу в списке по шаблону, в рез-тат не входит логин адреса
        result.append(domen[0]) # доб. в рез-т по первому эл(чтобы получить слитый список)
    for mail in set(result):
        print(mail+':', result.count(mail)) 
       
````

````python
count_email(emails)
````
