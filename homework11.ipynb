## задание1

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
def stat_rating(row):
    """Класифицирует статус рейтинга"""
    if (row['rating'] < 3):
        return 'низкий'
    if (row['rating'] == 4 and row['rating'] == 3):
        return 'средний'
    else:
        return 'высокий'
```

```python
ratings['rating_status'] = ratings.apply(stat_rating, axis=1)
```

```python
ratings[['userId', 'movieId', 'rating', 'rating_status']].head()
```

Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

Результат классификации запишите в отдельный столбец region.

```python
df = pd.read_csv('keywords.csv')
df.head(10)
```

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск'] }
```

```python
def geo_log2 (row):
    if (row['keyword'] in geo_data.values()):
        return geo_data['key']
    return 'undefined'
```

```python
df['region'] = df.apply(geo_log2, axis=1)
```

```python
def geo_log (row):
    """ Определяет регион поисковых запросов """
    if ('москва' in row['keyword']) and ('тула' in row['keyword']) and ('ярославль' in row['keyword']):
        return 'Центр'
    if ('петербург' in row['keyword']) and('петербург' in row['keyword']) and ('псков' in row['keyword']) and ('мурманск' in row['keyword']):
        return 'Северо-Запад'
    if ('владивосток' in row['keyword']) and ('сахалин' in row['keyword']) and('хабаровск' in row['keyword']):
        return 'Дальний Восток'
    else:
        return 'undefined'
```

```python
df['geo_data'] = df.apply(geo_log, axis=1)
```
