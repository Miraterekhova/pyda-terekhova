# выбросы

Будем исследовать метрики здоровых лошадей, поэтому выбросами будут какие-либо критичные значения разных параметров.
- температура 

Нормальная температура составляет 37,8, поэтому уберем все, что ниже 36,5 и все что выше 39,1 (с равным размахом 1,3).
- пульс

Убираем значения меньше 50 и больше 150
- слизистая оболочка

берем значени 1,2 и 3 (1,2 - с лошадью все хороше, значение 3 свидетельстует о ранем шоке)

- боль
1 = тревога, не боль

2 = депрессия

3 = перемежающаяся умеренная боль

4 = кратковременные сильные боли

5 = непрерывная сильная боль

Берем значения 1, 2, 3

- вздутие живота
1 = Нет

2 = незначительно

3 = умеренный

4 = тяжелый

Выбираем значения 1,2

- живот 

1 = нормальный

2 = другое

3 = твердые фекалии в толстой кишке

4 = растянутая тонкая кишка

5 = раздутая толстая кишка

т.к. 3 требует медикаментозного лечения без хирургического вмешательства, то выбираем значения от 1 до 3

```python
# создаем норвый дф применяя фильтры для удобства вычислений
healthy_house =pd.DataFrame()
```

```python
healthy_house['Age'] = horse_df['Age']
healthy_house['Hospital_number'] = horse_df['Hospital_number']
healthy_house['Temperature'] = horse_df.Temperature[(horse_df.Temperature >36.5) & (horse_df.Temperature < 39.1)]
healthy_house['Pulse'] = horse_df.Pulse[(horse_df.Pulse >50) & (horse_df.Pulse < 150)]
healthy_house['Mucous_membranes'] = horse_df.Mucous_membranes[(horse_df.Mucous_membranes == 1) |(horse_df.Mucous_membranes == 2)| (horse_df.Mucous_membranes == 3)]
healthy_house['Pain'] = horse_df.Pain[(horse_df.Pain == 1) | (horse_df.Pain == 2) | (horse_df.Pain == 3)]
healthy_house['Abdominal_distension'] = horse_df.Abdominal_distension[(horse_df.Abdominal_distension == 1) | (horse_df.Abdominal_distension == 2)]
healthy_house['Abdomen'] = horse_df.Abdomen[(horse_df.Abdomen >0) & (horse_df.Abdomen < 4)]
```

```python
healthy_house
```

видим, что кол-во записей не изменилось, но наши фильтры заполнились пустыми значениями, рассчитаем лосновные метрики для данного датафрейма и посмотрим как изменились показатели для здоровых лошадей

```python
print(healthy_house.median())
print(horse_df.median())
```

```python
print(healthy_house.std())
print(horse_df.std())
```

```python
print(healthy_house.var())
print(horse_df.var())
```

# Пропуски

```python
for col in horse_df.columns:
    missing = horse_df[col].isnull().mean()
    print(f'{col} - {missing :.1%}')
```

 Больше всего пропусков в колонке со вздутием живота, колонки возраста и номера не имеют пропусков

```python
# кол-во пропусков в каждой колонке
for col_count in horse_df.columns:
    missing_count = horse_df[col_count].isnull().sum()
    print(f'{col_count} - {missing_count}')
```

В колонках "Abdominal_distension", "Pain", "Temperature" логичнее пропуски заполнять медианой, т к в средне квадратичном отклонении показатель сильно изменилася при выбросах.

В колонке с пульсом самое незначительное кол-во пропусков, поэтому можно запонить колонку модой.

Т.к. показатель "Abdomen " сыграл важню роль в рассчете для показателей здоровых лошадей, его необходимо заполнить средним арифмитическим в разрезе возраста, чтобы вычисления были максимально точнные

колонку "Mucous_membranes" заполним среднем арифмитическим

```python
horse_df.Pulse.fillna(horse_df.Pulse.mode().loc[0], inplace=True)
horse_df.Abdominal_distension.fillna(round(horse_df.Abdominal_distension.median()), inplace=True)
horse_df.Pain.fillna(round(horse_df.Pain.median()), inplace=True)
horse_df.Temperature.fillna(horse_df.Temperature.median(), inplace=True)
horse_df.Mucous_membranes.fillna(horse_df.Temperature.mean(), inplace=True)
horse_df.Abdomen.fillna(horse_df.groupby(['Age', 'Abdominal_distension']).Abdomen.transform('mean'), inplace=True)
```
